#Escalamos los datos, si necesario


      Si es necesario porque hay datos que varian mucho por lo que usamos standard scaler para que la media sea cero y le SDV sea uno. Con estos resultados sabemos que los datos estan clusterizados cerca de la media. 

```python
#Importamos la librería StandardScaler
from sklearn.preprocessing import StandardScaler


#Creamos el objeto StandardScaler
scaler = StandardScaler()

#Estandarizamos los datos
dfScaler = scaler.fit_transform(df)

#Creamos el DataFrame con los datos estandarizados
dfFit = pd.DataFrame(dfScaler, columns = df.columns) 
#Mostramos los datos estandarizados
dfFit.head()
```

#Reduccion con PCA si necesario

      Es necesario reducir con PCA ya que hay muchas columnas y por lo tanto 
      mucha informacion. Lo hacemos con el siguiente codigo:

```python


#Importamos la librería PCA
from sklearn.decomposition import PCA

#Creamos el objeto PCA
pca = PCA()

#cambiamos nombres de columnas
pca_columns= ['PC1','PC2','PC3','PC4','PC5','PC6','PC7','PC8','PC9','PC10','PC11','PC12','PC13','PC14','PC15']

#Reducimos las dimensiones
dfPCA = pca.fit_transform(df)

#Creamos el DataFrame con los datos reducidos
dfFit = pd.DataFrame(dfPCA, columns=pca_columns)

#Mostramos los datos reducidos
dfFit
```

##Indica la varianza de los datos explicada por cada componente seleccionado. Para actividades de exploración de los datos la varianza > 70%

```python
#Calculamos la varianza de los datos explicada por cada componente seleccionado
print("La varianza de los datos explicada por cada componente seleccionado es: {}".format(pca.explained_variance_ratio_))
```

##Indica la importancia de las variables en cada componente

```python
print("La importancia de las variables en cada componente es: {}".format(pca.components_))
```

##Elabora los histogramas de los atributos para visualizar su distribución


```python
#Importamos la librería matplotlib
import matplotlib.pyplot as plt

#Creamos los histogramas
df.hist(bins = 25, figsize = (20,15))
plt.show()
```

```python
#gráficos boxplot
df.boxplot(figsize = (20,15))
plt.show()
```

```python
Explained_var=  np.round(pca.explained_variance_ratio_,4)*100
Acumulated_var= np.cumsum(pca.explained_variance_ratio_)*100

fig = plt.figure(figsize = (20, 10))
 
# creating the bar plot
plt.bar(pca_columns,Explained_var)
plt.plot(pca_columns,Acumulated_var, color='black', marker='.')

plt.xlabel("Principal Components")
plt.ylabel("Percentage")
plt.title("Variance distribution of the Principal Components")
plt.show()
```

      podemos observar que con el pc1 y el pc2 podemos explicar la mayoria de los datos

```python
df_pca_variance= pd.DataFrame()
df_pca_variance['Explained_var']=Explained_var
df_pca_variance['Acumulated_var']=Acumulated_var
df_pca_variance.index=pca_columns
df_pca_variance
```

      En esta tabla podemos observar mas claramente que con los primeros 3 componentes
      tenemos explicado el 95% de los datos.
