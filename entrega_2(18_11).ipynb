##IMPORTANT VARIABLES ANALYSIS

```python
binarias = Pipeline(steps= [('bin', OneHotEncoder())])
colbinarias = ['CUMPLE_CON_ALC', 'CUMPLE_CON_COND', 'CUMPLE_CON_SDT_salin', 'CUMPLE_CON_FLUO', 
          'CUMPLE_CON_DUR', 'CUMPLE_CON_CF', 'CUMPLE_CON_NO3', 'CUMPLE_CON_AS',   
          'CUMPLE_CON_CD', 'CUMPLE_CON_CR', 'CUMPLE_CON_HG', 'CUMPLE_CON_PB',       
          'CUMPLE_CON_MN', 'CUMPLE_CON_FE']

contin = Pipeline(steps=[('escalaNormal', StandardScaler())])
colcontin = ['LONGITUD', 'LATITUD', 'ALC_mg/L', 'CONDUCT_mS/cm', 'SDT_M_mg/L','FLUORUROS_mg/L', 'DUR_mg/L',
            'COLI_FEC_NMP/100_mL', 'N_NO3_mg/L', 'AS_TOT_mg/L', 'CD_TOT_mg/L',
            'CR_TOT_mg/L', 'HG_TOT_mg/L','PB_TOT_mg/L', 'MN_TOT_mg/L', 'FE_TOT_mg/L']

transform_columns = ColumnTransformer(transformers=[
                                        ('var_cont', contin, colcontin),
                                        ('var_bin', binarias, colbinarias)])
```

```python
le = LabelEncoder()
y = le.fit_transform(df['SEMAFORO'])
```

```python
X = df.drop('SEMAFORO', axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= 0.8, random_state= 42)

pipe_d = Pipeline(steps=[('variables', transform_columns),
                       ('modelo', RandomForestClassifier(random_state= 42))])

pipe_d.fit(X_train, y_train)

pipe_d[1][1].feature_importances_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= 0.8, random_state= 42)

continuas = Pipeline(steps=[('escalaNormal', StandardScaler())])
colcontinuas = ['FLUORUROS_mg/L', 'CONDUCT_mS/cm', 'MN_TOT_mg/L', 'N_NO3_mg/L', 
            'AS_TOT_mg/L', 'DUR_mg/L', 'FE_TOT_mg/L',
            'ALC_mg/L', 'CR_TOT_mg/L', 'LONGITUD', 'LATITUD']


catego = Pipeline(steps= [('bin', OneHotEncoder())])
colcatego = ['CUMPLE_CON_FLUO', 'CUMPLE_CON_COND',
              'CUMPLE_CON_MN', 'CUMPLE_CON_NO3', 'CUMPLE_CON_AS',
              'CUMPLE_CON_DUR', 'CUMPLE_CON_FE', 'CUMPLE_CON_ALC','CUMPLE_CON_CR']

transform_columns = ColumnTransformer(transformers=[
                                        ('var_cont', continuas, colcontinuas),
                                        ('var_bin', catego, colcatego)])
```

```python
modelos = [RandomForestClassifier(random_state= 42), DecisionTreeClassifier(random_state= 42)]
nombres = ['Random Forest', 'Decision Tree']
pipes = []

for i in range(len(modelos)):
    pipes.append(Pipeline(steps=[('variables', transform_columns),
                            ('modelo', modelos[i])]))

    pipes[i].fit(X_train, y_train)
    print(nombres[i]+" accuracy:")
    print(pipes[i].score(X_test, y_test))
    print("\n")
```

      Se peude ver que el mejor clasificador es el de random forest con .95
      en comparacion a decision tree con .911

      REPORTE DE CALSIFICACION

```python
target_names = ['class 0', 'class 1', 'class 2'] 
y_pred = pipes[0].predict(X_test)
print(classification_report(y_test, y_pred, target_names=target_names))
```

```python
cm = confusion_matrix(y_test, y_pred, labels=pipes[0][1].classes_)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,
                               display_labels=pipes[0][1].classes_)
disp.plot()
plt.grid(True)
plt.show()
```
