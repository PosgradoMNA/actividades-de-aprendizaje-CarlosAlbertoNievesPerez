###The Basics and array creation

```python
# a numpy array or ndarray is similar to a list. Its usually fixed in size and each element is of the same type, in this case, integers.
import numpy as np
a=[1,2,3,4,5]
x=np.array(a)
type(x)
```

```python
#find the size of the array
x.size
```

```python
#find the shape of the array//size of the array in each dimension
x.shape
```

```python
#type of data in array
x.dtype
```

```python
#find the mean of the array
x.mean()
```

```python
#number of array dimensions or the rank of the array
x.ndim
```

```python
#we can change the elements of the array as follows
x[0]=200
x
```

###Slicing methods

Like lists, we can slice the numpy array. Slicing in python means taking the elements from the given index to another given index.

We pass slice like this: [start:end].

We can select the elements from 1 to 3 and assign it to a new numpy array d as follows:

```python
y=x[0:3]
y
```

```python
#we can assign the corresponding indexes to new values as follows:
y[1:4]=100,400
y
```

```python
#we can also define the steps in slicing like so: [start:end:step]
print(x[0:6:2])
```
