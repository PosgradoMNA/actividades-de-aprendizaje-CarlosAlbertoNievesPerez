##NORMALIZACION

La normalización de datos es esencial para el correcto funcionamiento de una base de datos. Si una empresa trabaja con una base de datos con información que ha sido normalizada, esta será mucho más efectiva. Esto incluiría una serie de beneficios muy importantes para el buen desarrollo de un negocio:

Ahorro de tiempo al tener la información bien ordenada
Un 90% más de éxito en el funcionamiento de las bases de datos
Eliminación de datos duplicados
Optimización de los datos almacenados
Evitar errores en el procesamiento de los datos
Mostrar una imagen más profesional de cara a nuestros clientes

```python
def normalizacion(x):
  min=np.min(x)
  max=np.max(x)
  rango=max-min
  return [(a-min)/rango for a in x]

  ##esta es la formula de NORMALIZACION de min max
```

```python
normalizacion(varX)
 
```
